## Load Data

```python
data = pd.read_csv('./data/bitstampUSD_1-min_data_2012-01-01_to_2017-05-31.csv')
```

## Clean Data

Convert integer timestamp to datetime

```python
data['dt'] = data['Timestamp'].apply(lambda x: datetime.datetime.fromtimestamp(float(x)))
```

```python
data.head()
```

Inspect missing data...

```python
data['YYmm'] = data['dt'].apply(lambda x: str(x.year)+'-'+str(x.month))
```

```python
null_pcts = data.groupby('YYmm', as_index=False).agg({'Open':lambda x: x.isnull().sum()/(x.isnull().sum()+x.count())})
```
